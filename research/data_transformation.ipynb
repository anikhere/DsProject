```python

import os
from pathlib import Path
os.chdir(r'C:\Users\Dell\OneDrive\Desktop\GIT DEMO\aws\DsProject')
from dataclasses import dataclass
from pathlib import Path
import sys
from pathlib import Path

@dataclass
class DataTrans:
    root_dir: Path
    data_path: Path
```

```python
from src.ds.constants import *
from src.ds.utlis.common import Load_yaml,create_dir
```

```python
class ConfigManager:
    def __init__(
            self,
            config_filepath=CONFIG_FILE_PATH,
            params_filepath=PARAMS_FILE_PATH,
            schema_filepath=SCHEMA_FILE_PATH
            ):
        self.config = Load_yaml(config_filepath)
        self.yaml= Load_yaml(params_filepath)
        self.schema= Load_yaml(schema_filepath)
        create_dir([self.config.artifacts_root])
    def get_data_transform(self)-> DataTrans:
        config = self.config.data_transform
        create_dir([config.root_dir])
        dtc = DataTrans(
            root_dir=config.root_dir,
            data_path= config.data_path
        )
        

        return dtc
        
        
```

```python
import os 
from src.ds import logger
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
class actual_code:
    def __init__(self,config: DataTrans):
        self.config = config
    
    def Train_test(self):
        data = pd.read_csv(self.config.data_path)
        train,test= train_test_split(data)

        train.to_csv(os.path.join(self.config.root_dir,'train.csv'),index=False)
        test.to_csv(os.path.join(self.config.root_dir,'test.csv'),index=False)

        logger.info('splitted data')
        logger.info(train.shape)
        logger.info(test.shape)
        
```

```python
try:
    config= ConfigManager()
    datatc = config.get_data_transform()
    trans = actual_code(config=datatc)
    trans.Train_test()
except Exception as e:
    raise e
```
